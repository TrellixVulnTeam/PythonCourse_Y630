**Sorting** - Implement two types of sorting algorithms: Merge sort and bubble sort.

```python
## Merge sort
import random, math
new_list = list(range(101))
random.shuffle(new_list)
print(new_list)
                   
def merge_sort(l, left, right):
    if left < right:
        middle = left + round((right-left) / 2)
        print(left, middle, right)
        merge_sort(l, left, middle)
        merge_sort(l, middle+1, right)
        l = merge(l, left, middle, right)
        return l
    
def merge(l, left, middle, right):
    left_sorted = l[left:middle]
    right_sorted = l[middle+1: right+1]
    both_sorted = sort_lists(left_sorted, right_sorted)
    return both_sorted

def sort_lists(left, right):
    left_index = 0
    right_index = 0
    left_max = len(left)
    right_max = len(right)
    both_sorted = []
    print(left, right)
    for i in range(left_max + right_max):
        if left_index >= left_max:
            both_sorted.append(right[right_index])
            right_index += 1
        elif right_index >= right_max:
            both_sorted.append(left[left_index])
            left_index += 1
        elif left[left_index] < right[right_index]:     
            both_sorted.append(left[left_index])
            left_index += 1
        else:
            both_sorted.append(right[right_index])
            right_index += 1
    print("finished", both_sorted)
    return both_sorted

merge_sort(new_list, 0, len(new_list) - 1)
print(new_list)
```
